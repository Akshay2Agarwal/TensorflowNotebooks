```python
#Softmax regression

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt


#training data setup
x1_label0 = np.random.normal(1, 1, (100, 1))
x2_label0 = np.random.normal(1, 1, (100, 1))
x1_label1 = np.random.normal(3, 1, (100, 1))
x2_label1 = np.random.normal(4, 1, (100, 1))
x1_label2 = np.random.normal(7, 1, (100, 1))
x2_label2 = np.random.normal(8, 1, (100, 1))

xs_label0 = np.hstack((x1_label0, x2_label0))
xs_label1 = np.hstack((x1_label1, x2_label1))
xs_label2 = np.hstack((x1_label2, x2_label2))

xs = np.vstack((xs_label0, xs_label1, xs_label2))
labels = np.matrix([[1., 0., 0.]]*len(x1_label0)+[[0., 1., 0.]]*len(x1_label1)+[[0., 0., 1.]]*len(x1_label2))

arr = np.arange(xs.shape[0])
np.random.shuffle(arr)
xs = xs[arr, :]
labels = labels[arr, :]


#test data setup
x1_test_label0 = np.random.normal(2, 1, (20, 1))
x2_test_label0 = np.random.normal(1, 1, (20, 1))
x1_test_label1 = np.random.normal(3, 1, (20, 1))
x2_test_label1 = np.random.normal(5, 1, (20, 1))
x1_test_label2 = np.random.normal(9, 1, (20, 1))
x2_test_label2 = np.random.normal(8, 1, (20, 1))

xs_test_label0 = np.hstack((x1_test_label0, x2_test_label0))
xs_test_label1 = np.hstack((x1_test_label1, x2_test_label1))
xs_test_label2 = np.hstack((x1_test_label2, x2_test_label2))

xs_test = np.vstack((xs_test_label0, xs_test_label1, xs_test_label2))
labels_test = np.matrix([[1., 0., 0.]]*len(x1_test_label0)+[[0., 1., 0.]]*len(x1_test_label1)+[[0., 0., 1.]]*len(x1_test_label2))

arr_test = np.arange(xs_test.shape[0])
np.random.shuffle(arr_test)
xs_test = xs_test[arr_test, :]
labels_test = labels_test[arr_test, :]

train_size, num_features = xs.shape

learning_rate = 0.01
training_epochs = 1000
num_labels = 3
batch_size = 100

X = tf.placeholder("float", shape=[None, num_features])
Y = tf.placeholder("float", shape=[None, num_labels])
W = tf.Variable(tf.zeros([num_features, num_labels]))
b = tf.Variable(tf.zeros([num_labels]))

y_model = tf.nn.softmax(tf.matmul(X, W)+b)
#cost = -tf.reduce_sum(Y * tf.log(y_model))
cost = tf.reduce_mean(-Y*tf.log(y_model)-(1-Y)*tf.log(1-y_model))
#cost = -tf.reduce_sum(tf.log())

train_op = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
correct_predictions = tf.equal(tf.argmax(y_model, 1), tf.argmax(Y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_predictions, "float"))

with tf.Session() as sess:
    tf.global_variables_initializer().run()
    
    for step in range(training_epochs * train_size // batch_size):
        offset = (step * batch_size) % train_size
        batch_xs = xs[offset:(offset+batch_size), :]
        batch_labels = labels[offset:(offset+batch_size)]
        
        err, _ = sess.run([cost, train_op], feed_dict={X: batch_xs, Y: batch_labels})
        print(step, err)
    
    w_val = sess.run(W)
    print("Weight matrix", w_val)
    b_val = sess.run(b)
    print("Bias values", b_val)
    
    accuracy_val = accuracy.eval(feed_dict={X: xs_test, Y: labels_test})
    print("Test accuracy", accuracy_val)


```
